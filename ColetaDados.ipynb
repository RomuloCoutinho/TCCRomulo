```python
auth = tw.AppAuthHandler(consumerKey, consumerSecret)
api = tw.API(auth)
```

Parametros da pesquisa 

```python
search_words = 'pandemia' + '-filter:retweets'   
language = 'pt'
```

```python
tweets = tw.Cursor(api.search,
                   q=search_words,
                   lang = language,
                   tweet_mode="extended").items(10)            
```

```python
for tweet in tweets:
   print(tweet)
```

A busca por 'tweets' contÃ©m inÃºmeras informaÃ§Ãµes e atributos, porÃ©m, sÃ³ preciso do texto do tweet e do nome do usuario

```python
tweet_detalhes= [[tweet.user.screen_name, tweet.full_text] for tweet in tweets]
```

```python
tweet_df = pd.DataFrame(data=tweet_detalhes, columns=["user", "text"])
```
